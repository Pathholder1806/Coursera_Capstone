#### Importing all the libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

#### Reading the dataframe

```python
dataframe = pd.read_csv('toronto.csv')
```

```python
dataframe.sort_values(by="PostalCode", inplace=True)
```

```python
dataframe.reset_index(drop=True, inplace=True)
```

```python
dataframe
```

#### Reading the Geospatial cordinates csv file

```python
cord = pd.read_csv("Geospatial_Coordinates.csv")
```

```python
cord
```

```python
cord.sort_values(by="Postal Code", axis=0, inplace=True)
```

```python
cord
```

```python
cord.drop(columns="Postal Code", inplace=True)
```

```python
cord
```

#### Joining the two tables formed

```python
data_final = dataframe.join(other=cord, how='right')
```

#### Final Dataframe

```python
data_final
```

#### Saving this Dataframe

```python
data_final.to_csv('toronto_part2.csv', index=False)
```
